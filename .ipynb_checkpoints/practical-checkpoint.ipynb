# Detection of attack-targeted scans from the Apache HTTP Server access logs
This study seeks to obtain data which will help to address machine learning based malware research gaps. The specific objective of this study is to build a benchmark dataset for Windows operating system API calls of various malware. This is the first study to undertake metamorphic malware to build sequential API calls. It is hoped that this research will contribute to a deeper understanding of how metamorphic malware change their behavior (i.e. API calls) by adding meaningless opcodes with their own dissembler/assembler parts.

the objectives of this study can be summarized as follows:
* to detect vulnerability scans.
* to detect XSS and SQLI attacks.
* to examine access log files for detections.

# Attacks launched on servers

## XSS 
Cross Site Scripting attacks work by embedding script tags in
URLs/HTTP requests and enticing unsuspecting users to click on
them, ensuring that the malicious javascript gets executed on the
victim's machine. <p>These attacks leverage the trust between the
user and the server and the fact that there is no input/output
validation on the server to reject javascript or other active code
characters.</p>Simple attacks contain HTML tags.

```python
tags = re.compile(r'((\%3C)|<)((\%2F)|\/)*[a-z0-9\%]+((\%3E)|>)')
javaScript = re.compile(r'((\%3C)|<)((\%69)|i|(\%49))((\%6D)|m|(\%4D))((\%67)|g|(\%47))[^\n]+((\%3E)|>)')
key_words = re.compile(r'(javascript|vbscript|expression|applet|script|embed|object|iframe|frame|frameset)')
```

### XSS Acunetix

```python
xss_acunetix = []
z = 0
for n in range (0, len(acunetix_uri_df.URI)):
    x = tags.search(acunetix_uri_df.URI[n])
    y = javaScript.search(acunetix_uri_df.URI[n])
    m = key_words.search(acunetix_uri_df.URI[n])
    if x != None:
        # print(z,"->",acunetix_uri_df.URI[n])
        xss_acunetix.append(acunetix_uri_df.URI[n])
    elif y != None:
#         print()
#         print(z,"=>",acunetix_uri_df.URI[n])
        xss_acunetix.append(acunetix_uri_df.URI[n])
    elif m != None:
        # print(z,"==>",acunetix_uri_df.URI[n])
        xss_acunetix.append(acunetix_uri_df.URI[n])
    else:
        continue
    z+=1

# display(xss_acunetix)

XSSattack_list_acunetix = []
for n in range (0,len(xss_acunetix)):
    attack = acunetix_df[acunetix_df.Loc == xss_acunetix[n]]
    XSSattack_list_acunetix.append(attack)
XSSattack_acunetix = pd.concat(XSSattack_list_acunetix)
XSSattack_acunetix = XSSattack_acunetix.sort_values(by='Date')
XSSattack_acunetix = XSSattack_acunetix.drop_duplicates()
# display(XSSattack_acunetix)
# XSSattack_acunetix.head(5)
```

## SQL Injection

Code injection can be any type of code like SQL, LDAP, XPath,
XSLT, HTML, XML and OS command injection.<p>Here, we are focusing on the most
prevalent injection, the SQL injection. For SQL injections to
work, the attacker has to jump out of the original SQL statement.</p>
This is usually done by the singlequote
(') or the doubledash
().
The singlequote
acts as a delimiter for an SQL query; the
doubledash
is the comment character in Oracle and MS SQL.

```python
sq_dd = re.compile(r'(\')|(\%27)|(\-\-)|(#)|(\%23)')
eq_sc = re.compile(r'((\%3D)|(=))[^\n]*((\%27)|(\')|(\-\-)|(\%3B)|(;))')
OR = re.compile(r'\w*((\%27)|(\'))(\s|\+|\%20)*((\%6F)|o|(\%4F))((\%72)|r|(\%52))')
```

### SQLI acunetix

```python
SQLI_acunetix = []
z=0
for n in range (0, len(acunetix_uri_df.URI)):
    s = sq_dd.search(acunetix_uri_df.URI[n])
    e = eq_sc.search(acunetix_uri_df.URI[n])
    o = OR.search(acunetix_uri_df.URI[n])
    if s != None:
#         print(z,acunetix_uri_df.URI[n])
        SQLI_acunetix.append(acunetix_uri_df.URI[n])
    elif e != None:
#         print(z,"=>",acunetix_uri_df.URI[n])
        SQLI_acunetix.append(acunetix_uri_df.URI[n])
    elif o != None:
#         print(z,"==>",acunetix_uri_df.URI[n])
        SQLI_acunetix.append(acunetix_uri_df.URI[n])
    else:
        continue
    z+=1

# display(SQLI_acunetix)

SQLIattack_list_acunetix = []
for n in range (0,len(SQLI_acunetix)):
    attack = acunetix_df[acunetix_df.Loc == SQLI_acunetix[n]]
    SQLIattack_list_acunetix.append(attack)
SQLIattack_acunetix = pd.concat(SQLIattack_list_acunetix)
SQLIattack_acunetix = SQLIattack_acunetix.sort_values(by='Date')
SQLIattack_acunetix = SQLIattack_acunetix.drop_duplicates()
# display(SQLIattack_acunetix)
# SQLIattack_acunetix.head(5)
```
